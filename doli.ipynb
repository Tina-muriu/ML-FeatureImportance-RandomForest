```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer=load_breast_cancer()

X_train, X_test,y_train,y_test=train_test_split(cancer.data,cancer.target,stratify=cancer.target,random_state=42)

```

```python
random_classifier=RandomForestClassifier(n_estimators=200,max_features=6,max_depth=3,random_state=42)

random_classifier.fit(X_train, y_train)
model=GradientBoostingClassifier(random_state=0,learning_rate=0.01)
model.fit(X_train,y_train)

print("the accuracy of the training set:{:.2f}".format(random_classifier.score(X_train,y_train)))
print("the accuracy of the testing set:{:.2f}".format(random_classifier.score(X_test,y_test)))
```

```python
print("feature Importance:{}\n".format(random_classifier.feature_importances_))
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure(figsize=(25,15))
n_features=cancer.data.shape[1]
plt.barh(range(n_features),random_classifier.feature_importances_,align='center')
plt.yticks(np.arange(n_features),cancer.feature_names)
```
